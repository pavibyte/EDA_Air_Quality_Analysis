```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

Load DataSet

```python
file_path = r"C:\EDA_Air_Quality_Analysis\data\raw\air_quality_data.csv" 
df = pd.read_csv(file_path)
```

Display basic info

```python
display(df.info())
display(df.head())
```

Convert 'last_update' to datetime

```python
df['last_update'] = pd.to_datetime(df['last_update'])
```

Handling missing values

```python
missing_values = df.isnull().sum()
display(missing_values)
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

AQI Distribution

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['pollutant_avg'], bins=30, kde=True, color='blue')
plt.title('Distribution of Average Air Quality Index')
plt.xlabel('AQI')
plt.ylabel('Frequency')
plt.show()
```

Pollution Trends Over Time

```python
top_cities = df.groupby('city')['pollutant_avg'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_cities.values, y=top_cities.index, palette='Reds')
plt.title('Top 10 Most Polluted Cities')
plt.xlabel('Average AQI')
plt.ylabel('City')
plt.show()
```
